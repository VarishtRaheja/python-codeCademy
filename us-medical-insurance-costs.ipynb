# U.S. Medical Insurance Costs

```python
# Importing all packages here
import csv
```

```python
def filename(file):
    with open(file) as insurance_csv:
        insurance_file = csv.DictReader(insurance_csv)
        rows = []
        for row in insurance_file:
            rows.append(row)
    return rows
main_file = filename("insurance.csv")
print(main_file)
```

```python
# Data Analysis
"""
1. Find out the average age of the patients in the dataset.
2. Analyze where a majority of the individuals are from.
3. Look at the different costs between smokers vs. non-smokers.
4. Figure out what the average age is for someone who has at least one child in this dataset.
5. return the number of males vs. females counted in the dataset
6. find geographical location of the patients
7. return the average yearly medical charges of the patients
8. creating a dictionary that contains all patient information
"""
```

```python
# Data Analysis - 1
def avg_age(dataset):
    total_age = 0
    for age in dataset:
        total_age += int(age.get("age"))
    average_age = total_age/len(dataset)
    return "The average age is: " + str(round(average_age,2)) + " years"
avg_age(main_file)
```

```python
# Data Analysis - 2
# 2. Analyze where a majority of the individuals are from.
# for main_dict in main_file:
#     print(main_dict['region'])
def majority_individuals_region(dataset):
    unique_regions = dict()
    for keys in dataset:
        regions = (keys['region'])
        if regions not in unique_regions:
            unique_regions[regions] = 1
        else:
            unique_regions[regions] += 1
    return unique_regions

all_major_regions = majority_individuals_region(main_file)
majority = max(zip(all_major_regions.values(), all_major_regions.keys()))
print(majority[1], majority[0])
```

```python
# Data Analysis - 3
# 3. Look at the different costs between smokers vs. non-smokers.
def smoker_non_charges(datafile):
    smoker_charges = 0
    non_smoker_charges = 0
    for data_points in datafile:
        if data_points['smoker']=='yes':
            smoker_charges += float(data_points.get('charges'))/len(datafile)
        else:
            non_smoker_charges += float(data_points.get('charges'))/len(datafile)
    return smoker_charges, non_smoker_charges
calculating_charges = smoker_non_charges(main_file)

print("The average cost of insurance who smokes is: {}".format(round(calculating_charges[0],2)))
print("The average cost of insurance who does not smoke is: {}".format(round(calculating_charges[1],2)))
```

```python
# Data Analysis - 4
# 4. Figure out what the average age is for someone who has at least one child in this dataset.
child = 0
for i in main_file:
    if i['children']!=0:
        child += int(i.get('age'))/len(main_file)
print("The average is: {:.2f}".format(child))
```

```python
# Data analysis - 5
# 5. return the number of males vs. females counted in the dataset

def count(dataset):
    male_count = 0
    female_count = 0
    for no_of_ppl in dataset:
        sex = no_of_ppl.get("sex")
        if sex == 'male':
            male_count += 1
        else:
            female_count += 1
    return "The nunmber of males counted are: " + str(male_count) + " and the number of females counted are: " + str(female_count)
count(main_file)
```

```python
# Data analysis - 6
# 6. find geographical location of the patients

def unique(dataset):
    unique_region = []
    for regions in dataset:
        geo_regions = regions.get("region")
        if geo_regions not in unique_region:
            unique_region.append(geo_regions)
    return "Unique regions are: " + str(unique_region)
    
unique(main_file)
```

```python
# Data analysis - 7
# 7. return the average yearly medical charges of the patients
def average_yearly_medical_charges(dataset):
    total_charges = 0
    length_charges = []
    for yearly_charges in dataset:
        main_key = yearly_charges["charges"]
        length_charges.append(main_key)
        total_charges += float(main_key)
    average_medical_charge = total_charges/len(length_charges)
    return round(average_medical_charge,2)


average_yearly_medical_charges(main_file)
```

```python
# Data analysis - 8
# 8. creating a dictionary that contains all patient information

ages_list = [int(ages['age']) for ages in main_file]
sex_list = [(ages['sex']) for ages in main_file]
bmi_list = [float(ages['bmi']) for ages in main_file]
children_list = [int(ages['children']) for ages in main_file]
smoker_list = [(ages['smoker']) for ages in main_file]
region_list = [(ages['region']) for ages in main_file]
charges_list = [float(ages['charges']) for ages in main_file]
def create_dictionary(ages, sex, bmi, children, smoker, region, charges):
    dictionary = dict()
    for index in range(len(main_file)):
        dictionary[index+1] = {"age":ages[index], "sex":sex[index], "bmi":bmi[index], "children":children[index],
                              "smoker":smoker[index], "region":region[index], "charges":charges[index]}
    return dictionary
final_dictionary = create_dictionary(ages_list,sex_list,bmi_list,children_list,smoker_list,region_list,charges_list)
print(final_dictionary)
```
